```python
import pandas as pd
import json

```

```python
# Read the JSON file
with open('restaurants_list.json', 'r', encoding='utf-8') as json_file:
    json_data = json.load(json_file)

# Convert JSON data to a DataFrame
json_df = pd.DataFrame(json_data)
json_df.head()
```

```python
# Read the CSV file into a DataFrame
csv_df = pd.read_csv('restaurants_info.csv', sep=';')
csv_df.head(20)
```

```python
# Merge the DataFrames on 'objectid'
merged_df = pd.merge(json_df, csv_df, on='objectID')

# Convert the merged DataFrame back to a JSON-compatible format
merged_json = merged_df.to_dict(orient='records')

# Write the merged data to a new JSON file
with open('merged_data.json', 'w', encoding='utf-8') as output_file:
    json.dump(merged_json, output_file, indent=4)

print("Merged JSON data has been written to 'merged_data.json'")
```
